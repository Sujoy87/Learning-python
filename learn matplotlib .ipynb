# learn matplotlib 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
#scatterplot 
plt.figure(figsize=(7 , 5))

a=np.array([ 1 , 4 , 8 , 12 , 16])
b=np.array([55 , 60 , 65 , 70 , 75])

plt.scatter(a , b, marker='*' , color = 'b' )
plt.xlabel("A_value")
plt.ylabel("B value")
plt.show()
```

```python
#lineplot

plt.figure(figsize=(7 , 5))

a=np.array([ 1 , 4 , 8 , 12 , 16])
b=np.array([75 , 71 ,85 , 21 , 55])
           
plt.plot(a , b, color = "r")  
plt.xlabel("a_value")
plt.ylabel("y_value")
plt.title("lineplot")

plt.show()           
           
```

```python
#legend
py=[1,2,3,4,5,6,7]
a=[50,51,52,48,47,49,46]
b=[43,42,40,44,33,35,37]
c=[45,48,48,46,40,42,41]
plt.figure(figsize=(7 , 5))

plt.plot(py , a , label= 'a_result')
plt.plot(py , b , label= 'b_result')
plt.plot(py , c , label= 'c_result')

plt.legend(loc = 3)
plt.show()
```

```python
#histogram

a=np.array([ 1 , 15 , 8 , 7 , 16 , 25])
plt.figure(figsize=(5 , 3))
plt.hist(a)
plt.title('Histogram')
plt.show()
```

```python
plt.figure(figsize=(5 , 5))
py=[1,2,3,4,5,6,7]
a=[50,51,52,48,47,49,46]
b=[43,42,40,44,33,35,37]
c=[45,48,48,46,40,42,41]
plt.figure(figsize=(7 , 5))

plt.subplot(2 , 2 ,1)
plt.plot(py , a , label= 'a_result')
plt.plot(py , b , label= 'b_result')
plt.plot(py , c , label= 'c_result')
plt.title('all in one')

plt.subplot(2 , 2 , 2)
plt.plot(py , a , color="g")
plt.title('a')


plt.subplot(2 , 2 , 3)
plt.plot(py , b , color= 'b')
plt.title('b')

plt.subplot(2 , 2 , 4)
plt.plot(py , c , color = "r")
plt.title('c')



plt.show()
```

```python
#bar chart 
plt.figure(figsize=(7 , 5))
categories = ['a' , 'b' , 'c' , 'd']

values = np.array([25,50,75,30])

plt.bar(categories , values )
plt.grid()
plt.title('bar chart')
plt.xlabel('categories')
plt.ylabel('values')

plt.show()
```

```python
#pai chart 

languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#']
popularity = [30, 25, 15, 10, 20]
a = ['blue', 'orange', 'green', 'red', 'purple']


plt.pie(popularity , labels=languages , colors=a , autopct='%1.1f%%' , shadow= True)
plt.show
```

```python
#pai chart 

languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#']
popularity = [30, 25, 15, 10, 20]
a = ['blue', 'orange', 'green', 'red', 'purple']

explode=[0.5 , 0.4 , 0.3 , 0.2 , 0.1]
plt.pie(popularity , labels=languages , colors=a , autopct='%1.1f%%' , shadow= True , explode=explode)
plt.show
```
